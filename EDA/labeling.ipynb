```python
extracted=soup.find_all(['dt','b'])
#print (extracted)
extracted_text=[i.get_text() for i in extracted]
#print (extracted_text)
starting_index=extracted_text.index('1.0 Global and overview maps')
interest_list=extracted_text[starting_index:]
#print (interest_list)
test=['1.1 Carbohydrate metabolism', '00010', '00020', '00030', '00040', '00051', '00052', 
      '00053', '00500', '00520', '00620', '00630', '00640', '00650', '00660', '00562', 
      '1.2 Energy metabolism', '00190', '00195', '00196', '00710', '00720', '00680', '00910', '00920']
```

```python
label={};map_id={};map_name={}
for i,j in enumerate(interest_list):
    if '.' in j:
        name='-'.join(j.split()[1:])
        idd=j.split()[0]
        label[name]=[]
        for m,n in enumerate(interest_list[i+1:]):
            if '.' not in n:
                maps='map'+n
                label[name].append(maps)
                map_name[maps]=name
                map_id[maps]=idd
            else:
                break
            
print (label)
```

```python
pickle.dump(map_name, open( "mapname.pkl", "wb" ) )
pickle.dump(map_id, open( "mapid.pkl", "wb" ) )
pickle.dump(label, open( "label.pkl", "wb" ) )
```
